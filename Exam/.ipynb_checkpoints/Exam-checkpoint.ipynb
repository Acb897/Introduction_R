```r
GDP <- read.csv("Data/GDP.csv")
HDI <- read.csv("Data/HDI.csv")

head(GDP)
head(HDI)
str(GDP)
str(HDI)
```

```r
df <- cbind(GDP[order(GDP$Province), ], HDI[order(HDI$Province), ])
df <- df[, -4]
head(df)
tail(df)
```

```r
gdp_m <- aggregate(df$GDP.per.capita, list(df$CCAA), mean)
hdi_m <- aggregate(df$HDI, list(df$CCAA), mean)
gdp_sd <- aggregate(df$GDP.per.capita, list(df$CCAA), sd)
hdi_sd <- aggregate(df$HDI, list(df$CCAA), sd)
df_CCAA <- data.frame(CCAA = gdp_m$Group.1, GDP_m =  gdp_m$x, HDI_m = hdi_m$x, GDP_sd =  gdp_sd$x, HDI_sd = hdi_sd$x)
df_CCAA[is.na(df_CCAA)] <- 0
df_CCAA
```

```r
library(ggplot2)

# default plots size
options(repr.plot.width = 20, repr.plot.height = 10, repr.plot.res = 100)

# default theme and font size
theme_set(theme_bw(base_size = 24))


```

```r
ggplot(df, aes(x = GDP.per.capita, y = HDI)) + 
    geom_point(aes(color=CCAA), size=3)
```

```r
ggplot(df, aes(x = GDP.per.capita, y = HDI)) + geom_text(aes(label=Province, color=CCAA), size=8, check_overlap=T)
```

```r
ggplot(df, aes(x = GDP.per.capita, y = HDI)) + 
    geom_point(aes(color=CCAA), size=4) +
    geom_point(data=df_CCAA, aes(x = GDP_m, y = HDI_m), size=4) + 
    geom_errorbar(data=df_CCAA, aes(x = GDP_m, y = HDI_m, ymin = HDI_m-HDI_sd, ymax = HDI_m+HDI_sd)) +
    geom_errorbarh(data=df_CCAA, aes(y = HDI_m, x = GDP_m, xmin = GDP_m-GDP_sd, xmax = GDP_m+GDP_sd)) +
    geom_label(data=df_CCAA, aes(x = GDP_m, y = HDI_m, label=CCAA), size=6, hjust=1, vjust=1)
    
```
